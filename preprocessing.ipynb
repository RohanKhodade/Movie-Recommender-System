# transforming genres,keywords,overview,crew,cast into tags column

```python
movies.iloc[0]["genres"]
```

```python
# [{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
# into ["Action","Adventure","Fantasy","Science_Fiction"]
```

```python
import ast
# to convert str of list into list 
#ast.literal_eval(obj)
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i["name"])
    return L
```

```python
convert(movies.iloc[0]["genres"])
```

```python
movies["genres"]=movies["genres"].apply(convert)
```

```python
movies["keywords"]=movies["keywords"].apply(convert)
```

```python
movies.head()
```

```python
len(movies["cast"][0])
```

```python
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i["name"])
            counter+=1
        else:
            break
    return L
```

```python
movies["cast"].apply(convert3)
```

```python
movies["cast"]=movies["cast"].apply(convert3)
```

```python
movies.head()
```

```python
movies["crew"].iloc[0]
```

```python
def convertdir(director):
    L=[]
    for i in ast.literal_eval(director):
        if i["job"]=="Director" :
            L.append(i["name"])
            break
    return L
```

```python
movies["crew"].apply(convertdir)
```

```python
movies["crew"]=movies["crew"].apply(convertdir)
```

```python
movies.head()
```

```python
movies["overview"]=movies["overview"].apply(lambda X:X.split())
```

```python
movies.head()
```

```python
movies["genres"]=movies["genres"].apply(lambda x:[i.replace(" ","")for i in x])
movies["keywords"]=movies["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
movies["cast"]=movies["cast"].apply(lambda x:[i.replace(" ","")for i in x])
movies["overview"]=movies["overview"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

creating new column for tags

```python
movies["tags"]=movies["overview"]+movies["genres"]+movies["keywords"]+movies["cast"]+movies["crew"]
```

```python
movies["tags"]=movies["tags"].apply(lambda x:" ".join(x))
```

```python
movies.head()
```

```python
movies["tags"]=movies["tags"].apply(lambda x:x.lower())
```

```python
movies.head()
```

```python
new_df=movies[["movie_id","title","tags"]]
```
